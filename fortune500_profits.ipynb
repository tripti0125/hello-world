```python
non_numeric_profit = df['Profit (in millions)'].str.contains('[^0-9.-]')
df[non_numeric_profit]
```

```python
df[non_numeric_profit]['Profit (in millions)'].unique()
```

```python
df.columns = ['year', 'rank', 'company', 'revenue', 'profit']
```

```python
df.info()
```

```python
df = df.loc[~non_numeric_profit]
```

```python
df['profit'] = pd.to_numeric(df.profit)
df.dtypes
```

```python
group_by_year = df.groupby('year')['profit'].mean()
print(group_by_year)
```

```python
ax = plt.plot(group_by_year)
plt.title('Average Yearly Profits of Fortune500 Companies')
```
